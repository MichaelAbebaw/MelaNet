```python
from IPython.display import Audio, display
from src.data_generator import plot_spectrogram_feature
import librosa
import matplotlib.pyplot as plt
import numpy as np


def plot_raw(path, title):
    raw_audio = librosa.load(path)[0]
    fig = plt.figure(figsize=(12,2))
    steps = len(raw_audio)
    plt.plot(np.linspace(1, steps, steps), raw_audio)
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.show()

def plot_spectrogram_feature(spect):
    fig = plt.figure(figsize=(12,5))
    ax = fig.add_subplot(111)
    im = ax.imshow(spect, cmap=plt.cm.jet, aspect='auto')
    plt.title('Normalized Spectrogram')
    plt.ylabel('Time')
    plt.xlabel('Frequency')
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(im, cax=cax)
    plt.show()
    
display(Audio("speaker1_1.wav", embed=True))
display(Audio("speaker1_2.wav", embed=True))
display(Audio("speaker2_1.wav", embed=True))

plot_raw("speaker1_1.wav", "Speaker 1 record 1")
plot_raw("speaker1_2.wav", "Speaker 1 record 2")
plot_raw("speaker2_1.wav", "Speaker 2")


from src.data_generator import plot_spectrogram_feature

# plot normalized spectrogram
plot_spectrogram_feature(vis_spectrogram_feature)
# print shape of spectrogram
display(Markdown('**Shape of Spectrogram** : ' + str(vis_spectrogram_feature.shape)))
```
