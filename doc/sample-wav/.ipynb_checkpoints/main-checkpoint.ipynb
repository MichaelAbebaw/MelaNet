```python
from IPython.display import Audio, display
import librosa
import matplotlib.pyplot as plt
import numpy as np


def plot_raw(path):
    raw_audio = librosa.load(path)[0]
    fig = plt.figure(figsize=(12,3))
    steps = len(raw_audio)
    plt.plot(np.linspace(1, steps, steps), raw_audio)
    plt.title('Audio Signal')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.show()

display(Audio("speaker1_1.wav", embed=True))
display(Audio("speaker1_2.wav", embed=True))
display(Audio("speaker2_1.wav", embed=True))

plot_raw("speaker1_1.wav")
plot_raw("speaker1_2.wav")
plot_raw("speaker2_1.wav")
```
